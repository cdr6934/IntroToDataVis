```python
import pandas as pd
import numpy as np
import altair as alt
from vega_datasets import data
```

```python
mental_health_data = pd.read_csv( "Indicators_of_Anxiety_or_Depression_Based_on_Reported_Frequency_of_Symptoms_During_Last_7_Days.csv").query(
    'Group == "By State"')
```

```python
dataset = mental_health_data[(mental_health_data['Group'] == "By State") &
(mental_health_data['Indicator'] == "Symptoms of Anxiety Disorder") &
(mental_health_data['Time Period'] == 1)]
```

```python
pop = data.population_engineers_hurricanes() # simply to get the ids in order to merge the data
pop_id = pop[['state', 'id']]
```

```python
dataset = dataset.merge(pop_id, left_on='State', right_on='state')
dataset.head()
```

```python

states = alt.topo_feature(data.us_10m.url, 'states')

alt.Chart(states).mark_geoshape().encode(
    color='Value:Q',
    tooltip='Value:Q'
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(dataset, 'id', list(dataset.columns))
).properties(
    width=500,
    height=300
).project(
    type='albersUsa'
)
```

```python

states = alt.topo_feature(data.us_10m.url, 'states')
source = data.income.url

alt.Chart(source).mark_geoshape().encode(
    shape='geo:G',
    color='pct:Q',
    tooltip=['name:N', 'pct:Q'],
    facet=alt.Facet('group:N', columns=2),
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(data=states, key='id'),
    as_='geo'
).properties(
    width=300,
    height=175,
).project(
    type='albersUsa'
)
```
